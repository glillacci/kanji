## Convert to a Pandas data frame

```python
preproc_dict = {
    level: [c for c in SOURCE_DIC[level].split(' ') if c not in ['', '\n']]
    for level in SOURCE_DIC.keys()
}
```

```python
df = pd.DataFrame(columns=['kanji', 'kanken_level'])

for k, v in preproc_dict.items():
    row = {
        'kanken_level':  k,
        'kanji': v
    }
    df = df.append(row, ignore_index=True)

df = df.explode('kanji').reset_index(drop=True)
```

```python
df = df.groupby('kanji').agg({'kanken_level': 'max'}).reset_index()
```
